```python
import pandas as pd
import numpy as np

audio_set_df = pd.read_json('audioset_ontology.json')
```

```python
audio_set_labs = audio_set_df['name'].to_list()
```

```python
scapper_df = pd.read_csv('feat_ablation_sceneXevents.csv')
scapper_scenes = list(scapper_df['acoustic_scene_label'].unique())
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

language_model = SentenceTransformer('nli-roberta-base-v2')

def get_embeddings_similarity(label1, label2):
    emb1 = language_model.encode(label1)
    emb2 = language_model.encode(label2)
    return cosine_similarity([emb1], [emb2])[0][0]

```

```python
similarity_df = pd.DataFrame(columns=scapper_scenes, index=audio_set_labs)

for scene in scapper_scenes:
    for lab in audio_set_labs:
        similarity_df.at[lab, scene] = get_embeddings_similarity(scene, lab)
        print(scene, lab, similarity_df.at[lab, scene])
```
